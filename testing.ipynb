```python
import pandas as pd
from rag import rag
```

```python
df = pd.read_csv("questions.csv")
```

```python
num  = 0
file = 0
tot  = len(df['Question'])

import time

for i in range(tot):
    quest          = df['Question'][i]
    target_page    = df['Page'][i]
    target_document= df['Document'][i]

    seen_pages = set()
    seen_files = set()

    for src in rag(quest)['sources']:
        if str(src['pagenumber']) == str(target_page) and str(src['pagenumber']) not in seen_pages:
            num += 1
            seen_pages.add(src['pagenumber'])
        
        if src['filename'] == target_document and src['filename'] not in seen_files:
            file += 1
            seen_files.add(src['filename'])

    print(f"{num} for page number \n{file} for file\n")
    time.sleep(1)

print("Grade: " + str(num / tot))
print("Grade: " + str(file / tot))
```
